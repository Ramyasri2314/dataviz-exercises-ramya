# Week 2: Plotly Data Visualization

This exercise demonstrates basic visualizations using Plotly.

```python
import pandas as pd
import plotly.express as px
```

## Load Dataset

Using the built-in Gapminder dataset from Plotly.

```python
df = px.data.gapminder()
df.head()
```

## Line Chart

Life expectancy over years for different continents.

```python
fig = px.line(df, x="year", y="lifeExp", color="continent",
              title="Life Expectancy Over Time")
fig.show()
```

## Bar Chart

Population by continent for the year 2007.

```python
df_2007 = df[df["year"] == 2007]

fig = px.bar(df_2007, x="continent", y="pop", color="continent",
             title="Population by Continent (2007)")
fig.show()
```

## Scatter Plot

Relationship between GDP per capita and life expectancy.

```python
fig = px.scatter(df, x="gdpPercap", y="lifeExp",
                 color="continent",
                 title="GDP vs Life Expectancy")
fig.show()
```
